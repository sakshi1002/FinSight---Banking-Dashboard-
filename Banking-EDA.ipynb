Univariate Analysis


```python
bins = [0,100000,300000, float('inf')]
labels = ['Low','Mid','High']


df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels = labels, right=False)
```

```python
df['Income Band'].value_counts().sort_values(ascending=True).plot(kind='bar',color='skyblue',figsize=(8,5))
```

```python
categorical_cols = df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','IAId','Income Band']]

for col in categorical_cols:
    print(f"value counts for {col} : ")
    display(df[col].value_counts())
```

```python
for i, c in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=categorical_cols,x=c)
```

Bivariate Analysis


```python
for i, c in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=categorical_cols,x=c,hue='GenderId')
```

```python
for i, c in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=categorical_cols,x=c,hue='Nationality')
```

```python
#Histplot of value counte for different Occupation

for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

Numerical Analysis

```python
Numerical_cols = df[['Estimated Income','Superannuation Savings','Credit Card Balance']]


#Univariate Analysis and visualization

plt.figure(figsize=(25,20))
for i, n in enumerate(Numerical_cols):
    plt.subplot(3,1,i+1)
    sns.histplot(df[n],kde=True)
    plt.title(n)
    print()
plt.show()



    
```

```python
#heatmaps
Numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance']

correlation_matrix = df[Numerical_cols].corr()

plt.figure(figsize=(8,4))
sns.heatmap(correlation_matrix, annot=True, cmap='crest',fmt=".2f")
plt.show()
```
